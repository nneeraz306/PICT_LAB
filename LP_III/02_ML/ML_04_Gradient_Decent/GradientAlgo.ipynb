```python
import matplotlib.pyplot as plt

# Function and its derivative
def f(x):
    return (x + 3) ** 2

def df_dx(x):
    return 2 * (x + 3)
```

```python
x = float(input("Enter initial value for x (x₀): "))
steps = int(input("Enter number of steps: "))
step_size = float(input("Enter step size (learning rate): "))
```

```python
points = [(x, f(x))]  # store (x, f(x)) pairs for plotting

for i in range(steps):
    grad = df_dx(x)
    x_new = x - step_size * grad  # gradient descent formula

    print(f"Step {i+1}: x = {x_new:.4f}, f(x) = {f(x_new):.4f}, grad = {grad:.4f}")
    points.append((x_new, f(x_new)))

    # check if gradient is close to zero (convergence)
    if abs(grad) < 0.0001:
        print("\n✅ Converged: Gradient is close to zero.")
        break

    x = x_new  # update x

# Final output
print(f"\nLocal minimum found at x = {x:.4f}, f(x) = {f(x):.4f}")
```

```python
x_vals, y_vals = zip(*points)
plt.plot(x_vals, y_vals, marker='o')
plt.title("Gradient Descent Convergence")
plt.xlabel("x values")
plt.ylabel("f(x) = (x + 3)²")
plt.grid(True)
plt.show()
```
